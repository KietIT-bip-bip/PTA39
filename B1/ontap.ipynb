# 4 vòng lặp
- `while`
-`for` + `range(start, stop, step)`

```python
# while (vòng lặp vô hạn -> lặp đến khi điều kiện sai)
i = 0
while i < 7:
    if i == 6:
        continue # toi lan lap tiep theo
    if i == 3:
        pass # bo qua
    if i == 5:
        break # dung vong lap
    print(i)
    i += 1
```

```python
# range(stop) -> stop - 1
for i in range(10):
    print(i)
```

```python
# range(start, stop) -> stop - 1
for i in range(2, 10):
    print(i)
```

```python
# range(start, stop, step) -> stop - 1, step là bước nhảy (tính từ start)
for i in range(2, 10, 2): # chỉ lấy các giá trị chẵn (start: 2 + 2 lần bước nhảy)
    print(i)
```

# 7. Hàm (function)
- có trả về (`return`)
- có tham số truyền vào (`parameter`)

```python
# khai báo
def sumInList(your_list): # hàm truyền vào 1 tham số: your_list
    sum = 0 
    for item in your_list:
        sum += item
    return sum # hàm trả về giá trị của biến sum (1 giá trị cụ thể
```

```python
# gọi hàm
print(sumInList([12,0,0,1]))
```
